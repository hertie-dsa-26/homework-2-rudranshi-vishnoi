## Question 2: Pointers and References (1 point)

Examine the following Python code carefully:

```python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
```

What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

````python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
````

In Python, when you create a list, the variable doesn't actually store the list itself. It instead stores a reference to where the list is in memory. So original_array is just pointing to the list ["1_3", "5_2"] in memory.
When we pass original_array into the function, we are passing that same reference. So now both original_array and input (inside the function) are pointing to the exact same list.
The loop goes through each element and checks if the first character is "1". Since "1_3" starts with "1", it gets replaced with None. "5_2" starts with "5" so nothing happens to it.
Because both variables point to the same list in memory, when the function changes the list, original_array is also changed. So when we print both, they show the same result — [None, '5_2'].

## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

````python
import numpy as np

# Create a test array with numeric values
array = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
````

This algorithm is polynomial because it has two nested loops, both of which depend on the size of the input n. The total number of operations can be expressed as n(n+1)/2, which simplifies to ½n² + ½n. Since this can be expressed as a polynomial in terms of n, the algorithm is polynomial.

In Big-O notation we drop constants and smaller terms, so:
n(n+1)/2 
= ½n² + ½n
→ O(n²)
Therefore the runtime of this algorithm is O(n²).
